```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define alpha
alpha = 0.05

# Calculate critical values
z_alpha_2 = norm.ppf(alpha / 2)
z_1_minus_alpha_2 = norm.ppf(1 - alpha / 2)

# Generate x values and corresponding PDF values
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x)

# Plot the standard normal PDF
plt.plot(x, pdf, label='Standard Normal PDF')

# Shade the areas for alpha/2 and (1 - alpha/2)
plt.fill_between(x, 0, pdf, where=(x <= z_alpha_2), color='red', alpha=0.5, label=f'Alpha/2 = {alpha/2}')
plt.fill_between(x, 0, pdf, where=(x >= z_1_minus_alpha_2), color='blue', alpha=0.5, label=f'1 - Alpha/2 = {1 - alpha/2}')

# Define the ticks to display
ticks = np.concatenate((np.arange(-4, 5, 1), [z_alpha_2, z_1_minus_alpha_2]))
ticks = ticks[(ticks != -2) & (ticks != 2)]  # Remove -2 and +2 ticks

# Set the x-ticks
plt.xticks(ticks)

# Customize the colors of the specific ticks
for tick in plt.gca().get_xticklabels():
    tick_text = tick.get_text()
    try:
        tick_value = abs(float(tick_text))
        color = 'blue'
    except:
        tick_text = tick_text[1:]
        tick_value = abs(float(tick_text))
        color = 'red'
    if np.isclose(tick_value, z_1_minus_alpha_2, atol=0.01):
        tick.set_color(color)

# Rotate x-ticks
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('PDF of X ~ N(0,1)')
plt.legend()

# Save the plot
plt.savefig("images/standard_normal_pdf.png", dpi=300)

# Show the plot
plt.show()
```

```python
cdf = norm.cdf(x)
plt.plot(x, cdf, label='Standard Normal CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('CDF of X ~ N(0,1)')
plt.legend()
plt.savefig("images/standard_normal_cdf.png", dpi = 300) 
plt.show()
```
